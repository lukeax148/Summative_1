Next i will use the wbgapi library to pull data from the World Bank Data API and the pandas library to store the data in a dataframe.

```
def GDP_data(country, start_date, end_date):
    '''The GDP_data function will pull current Gross Domestic product data from the World Bank Data API
     depending on the country or countries selected (e.g. ['USA','CHN','GBR']) 
     and the start and end of the time series in years (e.g. 2000, 2010).
     
     >>> country = (['USA','GBR','CHN'])
     >>> x = GDP_data(country,2010,2022)     
     
     x.index.unique() = country
     
     '''
    data = wb.data.DataFrame('NY.GDP.MKTP.CN', country, range(start_date, end_date))
    return data
```
